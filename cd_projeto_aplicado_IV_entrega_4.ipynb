## Análise de Séries Temporais

Uma série temporal é um conjunto de observações registradas em intervalos sucessivos de tempo. Diferentemente de outros tipos de dados, as observações de uma série temporal apresentam dependência temporal, ou seja, os valores observados em determinado momento podem estar relacionados aos valores observados anteriormente.

A análise de séries temporais busca identificar padrões presentes nos dados, como tendências, sazonalidades e variações aleatórias. Esses padrões auxiliam na compreensão do comportamento dos fenômenos observados e fornecem informações importantes para análise e tomada de decisão.

Segundo Box et al. (2015), a análise temporal permite compreender a estrutura dos dados ao longo do tempo, identificando componentes que influenciam diretamente o comportamento da série.

## Componentes de uma Série Temporal

As séries temporais podem ser decompostas em diferentes componentes que ajudam a explicar seu comportamento.

### Tendência

A tendência representa a direção geral da série ao longo do tempo. Ela pode indicar crescimento, redução ou estabilidade do fenômeno analisado. No contexto da plataforma Memoria Litteraria, a tendência pode revelar o aumento ou diminuição da utilização da plataforma pelos usuários.

### Sazonalidade

A sazonalidade corresponde a padrões que se repetem em intervalos regulares de tempo. Esses padrões podem estar relacionados a períodos específicos, como dias da semana, meses ou eventos recorrentes que influenciam a utilização da plataforma.

### Componente Aleatório

O componente aleatório representa as variações imprevisíveis que não podem ser explicadas pela tendência ou pela sazonalidade. Essas oscilações são comuns em dados reais e refletem fatores externos ou eventos não previstos.

## Média Móvel

A média móvel é uma técnica amplamente utilizada em séries temporais para suavizar oscilações de curto prazo e destacar tendências de longo prazo. O método consiste em calcular a média dos valores observados dentro de uma janela temporal específica, reduzindo o impacto de flutuações momentâneas.

Neste projeto, a média móvel foi utilizada para facilitar a visualização da evolução dos acessos e interações registradas na plataforma ao longo do período analisado.

## Autocorrelação

A autocorrelação mede o grau de relação entre observações de uma mesma série temporal em diferentes instantes de tempo. Essa análise permite identificar se os valores observados em determinado período apresentam dependência em relação a períodos anteriores.

A presença de autocorrelação indica que existe um padrão temporal nos dados, contribuindo para a compreensão do comportamento dos usuários da plataforma.

## Pré-processamento dos dados

A etapa de pré-processamento teve como principal objetivo integrar as três tabelas relacionais em uma única base analítica.

Essa integração foi realizada utilizando as chaves estrangeiras presentes na tabela **tb_acesso**, permitindo associar cada registro de acesso às características do respectivo usuário e da obra acessada.

A base integrada constitui o conjunto de dados utilizado nas análises exploratórias e na construção da série temporal empregada nas etapas posteriores do trabalho.

```python
df = acessos.merge(
    usuarios,
    left_on="fk_usuario",
    right_on="id",
    suffixes=("", "_usuario")
)

df = df.merge(
    livros,
    left_on="fk_livro",
    right_on="id",
    suffixes=("", "_livro")
)

print(df.shape)

df.head()
```

```python
df.describe(include="all")
```

A inspeção inicial da base integrada permite verificar a consistência dos dados utilizados na pesquisa.

Não foram identificados valores nulos nos atributos empregados durante as análises, permitindo prosseguir para a etapa exploratória sem necessidade de imputação de dados.

Também foi observado que a estrutura relacional foi preservada durante a integração das tabelas, garantindo que cada acesso permaneça corretamente associado ao respectivo usuário e à obra correspondente.

## Construção da série temporal

Embora a base possua diversas variáveis demográficas e bibliográficas, o objetivo principal deste trabalho consiste na análise temporal do volume de utilização da plataforma.

Assim, cada registro da tabela **tb_acesso** foi agregado por data, produzindo uma série temporal univariada cujo valor observado corresponde ao número diário de acessos realizados na plataforma.

```python
serie = (
    df
    .groupby(df["dt_hora"].dt.date)
    .size()
    .reset_index(name="acessos")
)

serie.columns = ["data","acessos"]

serie.head()
```

As médias móveis constituem uma das técnicas mais utilizadas na análise exploratória de séries temporais.

O gráfico anterior apresenta a evolução temporal do número de acessos registrados na plataforma ao longo do período analisado.

### Discussão

Observa-se que as médias móveis reduzem significativamente as oscilações diárias presentes na série original.

A média móvel de 7 dias evidencia o comportamento semanal da plataforma, enquanto a média móvel de 30 dias fornece uma visão mais ampla da evolução dos acessos.

## Decomposição da Série Temporal

Uma série temporal pode ser interpretada como a combinação de diferentes componentes responsáveis por explicar seu comportamento ao longo do tempo.

Neste trabalho foi utilizada a decomposição aditiva, na qual a série observada é representada pela soma de três componentes principais:

- Tendência;
- Sazonalidade;
- Componente residual.

A decomposição permite compreender quais padrões estão presentes na série antes da aplicação de modelos preditivos.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposicao = seasonal_decompose(
    serie["acessos"],
    model="additive",
    period=7
)

fig = decomposicao.plot()

fig.set_size_inches(15,10)

plt.show()
```

### Discussão

A Figura anterior apresenta a decomposição da série temporal em seus principais componentes.

O primeiro gráfico corresponde à série original de acessos diários.

O segundo apresenta a tendência estimada, permitindo identificar possíveis movimentos de crescimento ou redução da utilização da plataforma ao longo do período estudado.

O terceiro gráfico evidencia o componente sazonal, indicando padrões que se repetem periodicamente.

Por fim, o componente residual representa a parcela da variabilidade que não pode ser explicada pela tendência nem pela sazonalidade.

A decomposição fornece uma visão inicial da estrutura temporal da série e auxilia na escolha dos modelos de previsão empregados nas etapas seguintes.

## Teste de Estacionaridade

Grande parte dos modelos clássicos de séries temporais pressupõe que a série analisada seja estacionária.

Uma série estacionária apresenta média, variância e estrutura de autocorrelação aproximadamente constantes ao longo do tempo.

Para verificar essa propriedade foi aplicado o teste Dickey-Fuller Aumentado (ADF).

```python
from statsmodels.tsa.stattools import adfuller

resultado = adfuller(serie["acessos"])

print(f"Estatística ADF : {resultado[0]:.4f}")
print(f"Valor-p         : {resultado[1]:.4f}")

if resultado[1] < 0.05:
    print("Conclusão: rejeita-se H0. A série pode ser considerada estacionária.")
else:
    print("Conclusão: não foi possível rejeitar H0. A série não é estacionária.")
```

### Discussão

O teste Dickey-Fuller possui como hipótese nula que a série apresenta raiz unitária, ou seja, não é estacionária.

Caso o valor-p seja inferior a 0,05, rejeita-se essa hipótese, indicando que a série pode ser considerada estacionária.

A análise da estacionaridade é uma etapa fundamental para a correta aplicação de modelos da família ARIMA, uma vez que tais modelos pressupõem estabilidade estatística da série ou sua obtenção por meio de diferenciação.

## Análise da Autocorrelação

Após verificar a estacionaridade da série, analisou-se a dependência temporal existente entre as observações por meio da Função de Autocorrelação (ACF).

Essa análise permite identificar se os valores observados permanecem correlacionados com valores registrados em períodos anteriores, fornecendo informações importantes para a construção do modelo preditivo.

```python
from statsmodels.graphics.tsaplots import plot_acf

fig = plt.figure(figsize=(12,5))

plot_acf(
    serie["acessos"],
    lags=30
)

plt.show()
```

### Discussão

A função de autocorrelação mede a intensidade da relação existente entre observações separadas por diferentes intervalos de tempo.

Coeficientes elevados para pequenas defasagens indicam persistência temporal, sugerindo que os valores observados em determinado dia permanecem relacionados aos dias anteriores.

Essa informação auxilia na escolha da estrutura autorregressiva do modelo ARIMA.
