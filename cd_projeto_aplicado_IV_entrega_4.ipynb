## Ciência de Dados

A Ciência de Dados é uma área multidisciplinar que combina estatística, matemática, programação e conhecimento de domínio para extrair informações relevantes a partir de grandes volumes de dados. Seu principal objetivo é transformar dados brutos em conhecimento útil para apoiar a tomada de decisões. Em ambientes digitais, a Ciência de Dados tem sido amplamente utilizada para compreender padrões de comportamento dos usuários, identificar tendências e gerar insights estratégicos.

No contexto deste projeto, a Ciência de Dados é utilizada para analisar os registros de interação dos usuários da plataforma Memoria Litteraria, permitindo compreender a dinâmica de utilização do sistema e identificar padrões de comportamento ao longo do tempo.

## Análise de Séries Temporais

Uma série temporal é um conjunto de observações registradas em intervalos sucessivos de tempo. Diferentemente de outros tipos de dados, as observações de uma série temporal apresentam dependência temporal, ou seja, os valores observados em determinado momento podem estar relacionados aos valores observados anteriormente.

A análise de séries temporais busca identificar padrões presentes nos dados, como tendências, sazonalidades e variações aleatórias. Esses padrões auxiliam na compreensão do comportamento dos fenômenos observados e fornecem informações importantes para análise e tomada de decisão.

Segundo Box et al. (2015), a análise temporal permite compreender a estrutura dos dados ao longo do tempo, identificando componentes que influenciam diretamente o comportamento da série.

## Componentes de uma Série Temporal

As séries temporais podem ser decompostas em diferentes componentes que ajudam a explicar seu comportamento.

### Tendência

A tendência representa a direção geral da série ao longo do tempo. Ela pode indicar crescimento, redução ou estabilidade do fenômeno analisado. No contexto da plataforma Memoria Litteraria, a tendência pode revelar o aumento ou diminuição da utilização da plataforma pelos usuários.

### Sazonalidade

A sazonalidade corresponde a padrões que se repetem em intervalos regulares de tempo. Esses padrões podem estar relacionados a períodos específicos, como dias da semana, meses ou eventos recorrentes que influenciam a utilização da plataforma.

### Componente Aleatório

O componente aleatório representa as variações imprevisíveis que não podem ser explicadas pela tendência ou pela sazonalidade. Essas oscilações são comuns em dados reais e refletem fatores externos ou eventos não previstos.

## Média Móvel

A média móvel é uma técnica amplamente utilizada em séries temporais para suavizar oscilações de curto prazo e destacar tendências de longo prazo. O método consiste em calcular a média dos valores observados dentro de uma janela temporal específica, reduzindo o impacto de flutuações momentâneas.

Neste projeto, a média móvel foi utilizada para facilitar a visualização da evolução dos acessos e interações registradas na plataforma ao longo do período analisado.

## Autocorrelação

A autocorrelação mede o grau de relação entre observações de uma mesma série temporal em diferentes instantes de tempo. Essa análise permite identificar se os valores observados em determinado período apresentam dependência em relação a períodos anteriores.

A presença de autocorrelação indica que existe um padrão temporal nos dados, contribuindo para a compreensão do comportamento dos usuários da plataforma.

## Plataformas Digitais e Análise Comportamental

A crescente utilização de plataformas digitais gera grandes volumes de dados relacionados às interações dos usuários. Esses dados podem ser analisados para compreender padrões de navegação, frequência de utilização, preferências e comportamento geral dos usuários.

# **EDA e Pré-processamento dos dados**

### 1. Base de Dados

Os dados utilizados simulam o comportamento de usuários brasileiros em uma plataforma digital de leitura, o Memoria Litteraria em sua versão web.

O banco é composto por três tabelas:

- **tb_usuario**: dados demográficos (idade, sexo)  
- **tb_livro**: características das obras (gênero, idioma)  
- **tb_acesso**: registros de interação. Um novo registro é inserido no banco de dados toda vez que um usuário interage com uma obra na plataforma.

A análise baseia-se na integração dessas três estruturas. Os dados encontram-se disponíveis em um arquivo .db, e as consultas realizadas foram feitas via SQLite, através do gerenciador DBeaver. Não houve necessidade de transformação dos dados ao longo do processo. Houve necessidade, contudo, de tratamento de alguns campos, originalmente setados como nulos ou com caracteres especiais, para um padrão com maior legibilidade e normalização.


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('database.db')

usuarios = pd.read_sql('select * from tb_usuario', conn)
livros = pd.read_sql('select * from tb_livro', conn)
acessos = pd.read_sql('select * from tb_acesso', conn)

df = acessos.merge(usuarios, left_on='fk_usuario', right_on='id') \
            .merge(livros, left_on='fk_livro', right_on='id')

df['faixa_etaria'] = df['idade'].apply(lambda x: 'Jovem' if x < 35 else 'Adulto')

df.head()
```


### 2. Análise dos Padrões de Leitura a partir dos Dados

A análise foi conduzida com base nos dados gerados nas tabelas de usuários (tb_usuario), livros (tb_livro) e acessos (tb_acesso). O objetivo foi verificar se os padrões esperados de comportamento emergem empiricamente dos dados, sem imposição externa de interpretações.

```python
pd.crosstab(df['faixa_etaria'], df['idioma'])
```


### 3. Frequência de acessos por faixa etária

A diferença no volume de acessos entre os grupos etários é bastante expressiva.

Usuários mais jovens realizam significativamente mais acessos ao sistema do que usuários mais velhos. Em média, o número de acessos por usuário jovem é aproximadamente três vezes maior do que o observado entre usuários mais velhos.

Esse resultado evidencia um **padrão forte de maior atividade** entre usuários jovens. É possível notar, a partir disto, que o usuário jovem possui um comportamento mais voltado à exploração, enquanto que usuários mais velhos tendem a manter-se por mais tempo interessado em um certo autor ou obra. Aprensentam, portanto, os segundos, maior grau de fidelidade, enquanto os primeiros, mais jovens, tendem a ser mais curiosos e experimentar livros e autores novos.


```python
df.groupby('faixa_etaria').size()
```


### 4. Diversidade de consumo (gêneros e obras)

Ao analisar a variedade de gêneros literários acessados por usuário, verifica-se que usuários mais jovens tendem a explorar um número ligeiramente maior de gêneros literários do que usuários mais velhos.

Contudo, a diferença observada é pequena. Ambos os grupos apresentam níveis relativamente próximos de diversidade de consumo.

Dessa forma, identifica-se uma **tendência leve à maior diversidade entre jovens**, mas não um comportamento claramente distinto entre os grupos.


```python
df.groupby('faixa_etaria')['genero'].nunique()
```


### 5. Relação entre idade e leitura em língua estrangeira

Os dados indicam que usuários mais jovens acessam mais frequentemente obras em línguas estrangeiras, especialmente em inglês.

Essa diferença é consistente com o padrão observado anteriormente e reforça a ideia de maior abertura dos usuários jovens a conteúdos internacionais.

Ainda assim, a presença de leitura em língua estrangeira também é significativa entre usuários mais velhos, o que indica que o comportamento não é exclusivo de um único grupo. Isto pode ser explicado, em grande medida, pela maior exposição dos mais jovens a conteúdos de origem, especialmente a partir de redes sociais, e também da escolarização bilíngue, padrão adotado nos últimos anos em muitas escolas nacionais.


```python
pd.crosstab(df['faixa_etaria'], df['idioma'])
```

### 6. Relação entre sexo e gênero literário
A análise da distribuição de gêneros literários por sexo revela a existência de algumas tendências:

Usuários do sexo feminino apresentam maior incidência de leitura de romances e best sellers. Já os usuários do sexo masculino apresentam maior incidência de leitura de ficção científica, romances policiais e obras clássicas.
Entretanto, essas diferenças não são tão expressivas. Ambos os grupos consomem uma variedade ampla de gêneros, e há sobreposição significativa entre os padrões de leitura.

Portanto, pode-se afirmar que existe uma tendência de preferência por gênero associada ao sexo, mas essa tendência é moderada e não determinística.

```python
pd.crosstab(df['sexo'], df['genero'])
```

# Modelos

## 6. Regressção Linear

A regressão linear é um modelo estatístico utilizado para analisar a relação entre uma variável independente e uma variável dependente, ajustando uma reta que melhor representa essa relação. Seu objetivo é identificar como mudanças na variável explicativa influenciam o valor da variável resposta.

No presente projeto, a regressão linear foi empregada para investigar a relação entre a idade dos usuários e o número de acessos realizados na plataforma. Considerando a idade como variável independente e a quantidade de acessos como variável dependente, o modelo permitiu identificar uma relação negativa entre essas variáveis, indicando que usuários mais jovens tendem a apresentar maior nível de atividade no sistema.

```python
from sklearn.linear_model import LinearRegression

acessos_por_usuario = df.groupby('fk_usuario').size().reset_index(name='acessos')
dados_modelo = acessos_por_usuario.merge(usuarios, left_on='fk_usuario', right_on='id')

X = dados_modelo[['idade']]
y = dados_modelo['acessos']

modelo = LinearRegression()
modelo.fit(X, y)

modelo.coef_, modelo.intercept_
```
